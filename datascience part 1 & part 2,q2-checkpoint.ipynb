**1.1 Extracting Biographies**

```python
def get_bibliography(category: str, rdf_type: str, max_biographies: int ) -> List[str]:
    """
    Retrieve biographies from Wikipedia for a specified category and RDF type.

    Args:
        category (str): The category name (e.g., "Astronauts", "Chefs").
        rdf_type (str): The RDF type of the category (e.g., "dbo:Astronaut", "dbo:Chef").
        max_biographies (int): Maximum number of biographies to collect (default: 100).

    Returns:
        List[str]: A list of biographies extracted from Wikipedia.
    """
    biographies = []

    encoded_category = quote(category)  # Encode the category name for use in the SPARQL query

    sparql = SPARQLWrapper("http://dbpedia.org/sparql")
    query = f"""
        PREFIX dbo: <http://dbpedia.org/ontology/>
        PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>

        SELECT DISTINCT ?page ?summary
        WHERE {{
            ?page rdf:type {rdf_type} .
            ?page rdfs:label ?label .
            ?page dbo:abstract ?summary .
            FILTER (LANG(?label) = 'en') .
            FILTER (LANG(?summary) = 'en') .
        }}
        LIMIT {max_biographies}
    """

    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)

    results = sparql.query().convert()

    for result in results["results"]["bindings"]:
        biography = result["summary"]["value"]
        biographies.append(biography)

    return biographies
```

**1.2 Extracting Knowledge Graph Facts**

```python
def extract_kg_facts(person_names: List[str]) -> Dict[str, List[Dict]]:
    """
    Extract KG Graph facts (RDF triples) for the specified people from DBpedia.

    Args:
        person_names (List[str]): List of person names.

    Returns:
        Dict[str, List[Dict]]: A dictionary containing KG Graph facts for each person.
    """
    kg_facts = {}

    for person_name in person_names:
        kg_facts[person_name] = []

        sparql = SPARQLWrapper("http://dbpedia.org/sparql")
        encoded_person_name = quote(person_name)
        query = f"""
            PREFIX dbo: <http://dbpedia.org/ontology/>
            PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
            SELECT DISTINCT ?predicate ?object
            WHERE {{
                <http://dbpedia.org/resource/{encoded_person_name}> ?predicate ?object.
                FILTER (strstarts(str(?predicate), str(dbo:)))
            }}
        """

        sparql.setQuery(query)
        sparql.setReturnFormat(JSON)

        results = sparql.query().convert()

        for result in results["results"]["bindings"]:
            predicate = result["predicate"]["value"]
            obj = result["object"]["value"]
            kg_facts[person_name].append({"predicate": predicate, "object": obj})

    return kg_facts
```

```python
def get_bibliography_and_kg_facts(categories: List[Tuple[str, str]], max_biographies: int = 100) -> Dict[str, Dict[str, List[Dict]]]:
    """
    Retrieve biographies from Wikipedia for specified categories and their RDF types,
    and extract KG Graph facts (RDF triples) for each person mentioned in the biographies.

    Args:
        categories (List[Tuple[str, str]]): List of tuples containing category name and RDF type.
        max_biographies (int): Maximum number of biographies to collect for each category (default: 100).

    Returns:
        Dict[str, Dict[str, List[Dict]]]: A dictionary containing biographies and KG Graph facts for each category.
    """
    data = {}

    for category, rdf_type in categories:
        biographies = get_bibliography(category, rdf_type, max_biographies=max_biographies)
        person_names = [biography.split()[0] for biography in biographies]  # Extract person names from biographies
        kg_facts = extract_kg_facts(person_names)
        data[category] = {"biographies": biographies, "kg_facts": kg_facts}

    return data
```


   **1.3 Storing Data**
   - Biographies will be stored in a Pandas DataFrame, where each row represents a biography and its associated category.
   - Each biography will also be saved in a separate text file named **`person_category.txt`**.

```python
def store_biographies_as_dataframe(categories: List[Tuple[str, str]], max_biographies: int) -> pd.DataFrame:
    """
    Store each biography along with its category in a Pandas DataFrame.

    Args:
        categories (List[Tuple[str, str]]): List of tuples containing category names and RDF types.
        max_biographies (int): Maximum number of biographies to collect.

    Returns:
        pd.DataFrame: DataFrame containing biographies and categories.
    """
    biographies_list = []
    categories_list = []

    for category, rdf_type in categories:
        biographies = get_bibliography(category, rdf_type, max_biographies)
        biographies_list.extend(biographies)
        categories_list.extend([category] * len(biographies))

    # Create DataFrame
    df = pd.DataFrame({"Biography": biographies_list, "Category": categories_list})
    return df
```

```python
def store_biographies_as_text_files(categories: List[Tuple[str, str]], max_biographies: int):
    """
    Store each biography in a separate text file named person_category.txt.

    Args:
        categories (List[Tuple[str, str]]): List of tuples containing category names and RDF types.
        max_biographies (int): Maximum number of biographies to collect.
    """
    for category, rdf_type in categories:
        biographies = get_bibliography(category, rdf_type, max_biographies)
        for biography in biographies:
            # Store biography in a text file named person_category.txt
            file_name = f"{biography[:20]}_{category.replace(' ', '_')}.txt"
            with open(file_name, "w") as file:
                file.write(biography)
            print(f"File '{file_name}' created successfully.")
```

### Preprocessing Function (`preprocess_text`):
1. Normalize each sentence to lowercase.
2. Remove stopwords .

```python
biographies_df = df


def preprocess_text(text):
    doc = nlp(text)  # Tokenize
    sentences = [sent.text for sent in doc.sents]  # Segment into sentences
    return sentences


biographies_df['preprocessed_sentences'] = biographies_df['Biography'].apply(preprocess_text)

def preprocess_tokens(text):
    doc = nlp(text.lower())  # Tokenize and normalize to lowercase
    # Remove stopwords, punctuation, and non-alphabetic tokens
    preprocessed_text = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct and token.is_alpha]
    return preprocessed_text

biographies_df['preprocessed_text'] = biographies_df['Biography'].apply(preprocess_tokens)
```

```python
def collect_bigrams(tokens):
    bigrams = nltk.bigrams(tokens)
    occurences = {}
    for bigram in bigrams:
        if occurences.get(bigram):
            occurences[bigram] += 1
        else:
            occurences[bigram] = 1
    return occurences
```

* Writing a function that checks a biography for the tokens

1.   Checks the number of spans where there is complete overlap

2.   Partial overlap in spans predicted
3. Number of spans a package predicted as an NE, but other package did not
4. Spans with full and partial agreement



```python
import spacy
import stanza
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


import spacy
import stanza

def check_biography(biography):
    """
    Checks a biography for named entity recognition agreement between spaCy and Stanza.

    Args:
        biography (str): The biography to check.

    Returns:
        dict: A dictionary containing the following information:
            - full_agreement: Number of spans with complete overlap in predicted NE.
            - partial_agreement: Number of spans with partial overlap in predicted NE.
            - spacy_only: Number of spans predicted as NE only by spaCy.
            - stanza_only: Number of spans predicted as NE only by Stanza.
            - type_agreement: Number of spans with agreement in NE type.
    """
    # Process the biography with spaCy and Stanza
    doc_spacy = nlp_spacy(biography)
    doc_stanza = nlp_stanza(biography)

    # Extract entities
    spacy_entities = [(ent.start_char, ent.end_char, ent.label_) for ent in doc_spacy.ents]
    stanza_entities = [(ent.start_char, ent.end_char, ent.type) for sent in doc_stanza.sentences for ent in sent.ents]

    # Initialize counters
    full_agreement = 0
    partial_agreement = 0
    spacy_only = 0
    stanza_only = 0
    type_agreement = 0

    # Iterate over the spaCy spans
    for span_spacy in spacy_entities:
        full_overlap = False
        partial_overlap = False
        for span_stanza in stanza_entities:

            # Full agreement
            if span_spacy[0] == span_stanza[0] and span_spacy[1] == span_stanza[1]:
                full_agreement += 1
                full_overlap = True

                # Checking for type agreement
                if span_spacy[2] == span_stanza[2]:
                    type_agreement += 1
                break

            # Partial agreement
            elif (span_stanza[0] <= span_spacy[0] <= span_stanza[1]) or (span_stanza[0] <= span_spacy[1] <= span_stanza[1]):
                partial_agreement += 1
                partial_overlap = True
                break

        if not full_overlap and not partial_overlap:
            spacy_only += 1

    # Count Stanza-only spans
    for span_stanza in stanza_entities:
        if not any((span_stanza[0] <= span[0] <= span_stanza[1]) or (span_stanza[0] <= span[1] <= span_stanza[1]) for span in spacy_entities):
            stanza_only += 1

    return {
        "full_agreement": full_agreement,
        "partial_agreement": partial_agreement,
        "spacy_only": spacy_only,
        "stanza_only": stanza_only,
        "type_agreement": type_agreement,
    }
```

```python
#Iterating through the biographies to obtain the results
df = pd.DataFrame(biographies_df)
biographies_astronauts = df[df["Category"] == "Astronauts"]["Biography"].tolist()
biographies_chefs = df[df["Category"] == "Chefs"]["Biography"].tolist()

results_astronauts = [check_biography(biography) for biography in biographies_astronauts]
results_chefs = [check_biography(biography) for biography in biographies_chefs]
```

* Agregating the results we had for Spacy and Stanza for both Chefs and Astronauts






```python
def aggregate_counts(astronauts_data, chefs_data):
    # Combined totals
    total_full_agreement = astronauts_data['full_agreement'] + chefs_data['full_agreement']
    total_partial_agreement = astronauts_data['partial_agreement'] + chefs_data['partial_agreement']
    total_spacy_only = astronauts_data['spacy_only'] + chefs_data['spacy_only']
    total_stanza_only = astronauts_data['stanza_only'] + chefs_data['stanza_only']
    total_type_agreement = astronauts_data['type_agreement'] + chefs_data['type_agreement']

    # Return aggregated counts as a dictionary
    aggregated_counts = {
        'full_agreement': total_full_agreement,
        'partial_agreement': total_partial_agreement,
        'spacy_only': total_spacy_only,
        'stanza_only': total_stanza_only,
        'type_agreement': total_type_agreement
    }

    return aggregated_counts

#Results gathered for Astronauts
astronauts_data = {
    'full_agreement': 860,
    'partial_agreement': 168,
    'spacy_only': 23,
    'stanza_only': 79,
    'type_agreement': 1261
}
#results gathered for Chefs
chefs_data = {
    'full_agreement': 695,
    'partial_agreement': 82,
    'spacy_only': 9,
    'stanza_only': 65,
    'type_agreement': 988
}

aggregated_data = aggregate_counts(astronauts_data, chefs_data)

#print results
print("Astronauts Data:")
print(astronauts_data)
print("Chefs Data:")
print(chefs_data)

```
